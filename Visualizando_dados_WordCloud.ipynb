```
#gera o grafico na lina
#%matplotlib inline

import pandas as pd
from wordcloud import WordCloud
```

```
#base de palavras
textos = [
          'eu nao gostei disso', 
          'eu nao quero isso', 
          'nao vou mais aceitar isso',
          'eu nao gosto disso',
          'gosto disso eu agora',
          'voce goataria de que ir conosco'
          ]
resenha = pd.DataFrame(textos, columns=['texto'])
resenha
```

```
# compreens√£o de lista - https://www.alura.com.br/artigos/simplicando-o-processamento-com-compreensao-de-lista-do-python
todas_palavras = ' '.join([texto for texto in resenha.texto])

#todas_palavras[:3]
print('quantidade de palavras: ',len(todas_palavras))
```

```
#contruir as palavra
nuvem_palavra = WordCloud().generate(todas_palavras)
nuvem_palavra
```

```
#tranformar o objeto em visual
import matplotlib.pyplot as plt
```

```
plt.figure()
plt.imshow(nuvem_palavra)
plt.show()
```

```
#mudar o tamanho da distribuicao
nuvem_palavra = WordCloud(width=800, height=500).generate(todas_palavras)

plt.figure()
plt.imshow(nuvem_palavra)
plt.show()
```

```
#definir o tamanho maximo para as palavras mudando a escala
nuvem_palavra = WordCloud(width=800, height=500, max_font_size=110).generate(todas_palavras)

plt.figure()
plt.imshow(nuvem_palavra)
plt.show()
```

```
nuvem_palavra = WordCloud(width=800, 
                          height=500, 
                          max_font_size=110).generate(todas_palavras)

#tamanho da imagem
plt.figure(figsize=(10,7))
plt.imshow(nuvem_palavra)
plt.show()
```

```
nuvem_palavra = WordCloud(width=800, 
                          height=500, 
                          max_font_size=110).generate(todas_palavras)

plt.figure(figsize=(10,7))
#deixar as imagens mais nitidas
plt.imshow(nuvem_palavra, interpolation='bilinear')
plt.show()
```

```
#collocation calculo da frequancia pela palavra e nao pelo bigrama
nuvem_palavra = WordCloud(width=800, 
                          height=500, 
                          max_font_size=110,
                          collocations = False
                          ).generate(todas_palavras)

plt.figure(figsize=(10,7))
plt.imshow(nuvem_palavra, interpolation='bilinear')
plt.axis('off')
plt.show()
```
